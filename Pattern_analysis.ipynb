```python
def charts(data, folder, different_colors = None):
    x_names = ['2015', '2016', '2017', '2018', '2019', '2020', '2021']
    # x_names = [2015, 2016, 2017, 2018, 2019, 2020, 2021]

    for i in sorted(np.unique(data['pattern'].values)):
        fig, ax = plt.subplots()
        name = 'pattern ' + str(int(i))
        plt.title(name)

        vecs = data[data['pattern'] == i][x_names].values
        print(name)
        handles=data[data['pattern'] == i]['AfN'].values.tolist()
        for v in range(len(vecs)):
            if different_colors is None or i not in different_colors:
                ax.plot(x_names, vecs[v], c='b', label=handles[v])
            else:
                # ax.set_prop_cycle('color', sns.color_palette("magma", 1))
                ax.plot(x_names, vecs[v], label=handles[v])
                
        s = f'there are {len(vecs)}  vectors in the patterns'
        ax.text(0, -0.1, s, size='x-large')
        # ax.plot(vecs.mean(axis=0), linestyle = '--', c='r', label='mean vector')
        ax.axis(ymin=0,ymax=1)
        if len(handles) < 100:
            # Shrink current axis by 20%
            d = 20
            r = d/100
            box = ax.get_position()
            ax.set_position([box.x0, box.y0, box.width * (1 - r), box.height])

            # Put a legend to the right of the current axis
            ax.legend(loc='center left', bbox_to_anchor=(1, 0, r, 1))
        
        way = folder + '/' + name +'.png'
        fig.set_size_inches(18.5, 10.5)
        fig.savefig(way)
        fig.clf()
    plt.show()
    clear_output()
```
